```python
## Fetching integer gap for warmstart and model dependent
def get_integer_gap(intance_log):
    results = glt.parse([intance_log])
    summary = results.summary()
    #gets the value of the cell for summary['MIPGap']
    return summary.iloc[0]['MIPGap']
get_integer_gap('/Users/letshopethisworks2/Documents/Phd Paper material/model_runs/results_jan10/2024_01_07new_equip_small_seq7/md/model_dependent_linear/lmd_md_small0.log')

def read_csv_add_gap(df, gapname= 'gap'):
    df[gapname] = df['log'].apply(get_integer_gap)*100
    return df

def read_csv_add_log(log_file_location, file_name_stuff, df):
    df['log'] = df['group_counter'].apply(lambda x: log_file_location+ f'{file_name_stuff}{x}.log')
    return df

md_df = pd.read_csv('/Users/letshopethisworks2/Documents/Phd Paper material/model_runs/results_jan10/2024_01_07new_equip_small_seq7/md/model_dependent_linear/lmd_md_small_results.csv')

md_df = read_csv_add_log('/Users/letshopethisworks2/Documents/Phd Paper material/model_runs/results_jan10/2024_01_07new_equip_small_seq7/md/model_dependent_linear/', 'lmd_md_small', md_df)
md_df = read_csv_add_gap(md_df, gapname = 'gap_md' )
md_df.tail()
```
